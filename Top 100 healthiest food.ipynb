```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
df=pd.read_csv("Top 100 healthiest nutrition and protien food.csv")
```

```python
df.isnull().sum()
```

```python
df["Food"].unique()
```

```python
df["Food"]=label_encoder.fit_transform(df["Food"])
df["Food"].unique()
```

```python
df["Nutrition Value (per 100g)"].unique()
```

```python
df["Nutrition Value (per 100g)"]=label_encoder.fit_transform(df["Nutrition Value (per 100g)"])
df["Nutrition Value (per 100g)"].unique()
```

```python
df["Quantity"].unique()
```

```python
df["Quantity"]=label_encoder.fit_transform(df["Quantity"])
df["Quantity"].unique()
```

```python
df["Originated From"].unique()
```

```python
df["Originated From"]=label_encoder.fit_transform(df["Originated From"])
df["Originated From"].unique()
```

```python
df
```

```python
x=df.drop(["Food"],axis=1)
x
```

```python
y=df["Food"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=80)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_hat=model.predict(x_test)
```

```python
model.score(x,y)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_hat,y_test)
```
